```python
import cv2
import numpy as np

def Img_PCA(img,a):           #img为输入图像，a为保留主成分比例
    U,S,V=np.linalg.svd(img)  #将img进行奇异值分解，U为左酉矩阵大小M*M，S为奇异值，V为右酉矩阵N*N。
    SS=np.zeros(img.shape)
    for i in range(len(S)):
        SS[i][i]=S[i]       #将S转化为M*N对角矩阵

    sval=np.sum(S)
    for i in range(len(S)):
        if np.sum(S[:i]) >= a*sval:
            k = i + 1
            break
    Uk=U[:,0:k]
    Sk=SS[0:k,0:k]
    Vk=V[0:k,:]
    print(Uk.shape)
    print(Sk.shape)
    print(Vk.shape)
    img_pca = np.uint8(np.dot(np.dot(Uk,Sk),Vk))  #计算降维处理后的矩阵
    return img_pca

if  __name__ == "__main__":
    img = cv2.imread('0.jpg',0)
    img_pca = Img_PCA(img,0.8)
    cv2.imshow("img_gray",img)
    cv2.imshow("img_pca",img_pca)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import cv2
def is_inside(o, i):
    ox, oy, ow, oh = o
    ix, iy, iw, ih = i
    # 如果符合条件，返回True，否则返回False
    return ox > ix and oy > iy and ox + ow < ix + iw and oy + oh < iy + ih

# 根据坐标画出人物所在的位置
def draw_person(img, person):
  x, y, w, h = person
  cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 255), 2)

# 定义HOG特征+SVM分类器
img = cv2.imread("2.jpg")
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
found, w = hog.detectMultiScale(img, winStride=(8, 8), scale=1.05)

# 判断坐标位置是否有重叠
found_filtered = []
for ri, r in enumerate(found):
    for qi, q in enumerate(found):
        a = is_inside(r, q)
        if ri != qi and a:
            break
    else:
        found_filtered.append(r)
# 勾画筛选后的坐标位置
for person in found_filtered:
    draw_person(img, person)
# 显示图像
cv2.imshow("people detection", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import os
import cv2
import tkinter
import tkinter.filedialog
import numpy as np
import pixellib
from PIL import Image,ImageTk
from torchvision import transforms as transforms
from pixellib.semantic import semantic_segmentation
from pixellib.instance import instance_segmentation

#设置图片保存路径
outfile = './save'

#创建一个界面窗口
win = tkinter.Tk()
win.title("综合实践图像处理")
win.resizable(width=False, height=False)
screenwidth = win.winfo_screenwidth()
screenheight = win.winfo_screenheight()
size = '%dx%d+%d+%d' % (1073, 690, (screenwidth - 1073) / 2, (screenheight - 690) / 2)
win.geometry(size)

#设置全局变量
original = Image.new('RGB', (480, 480))
save_img = Image.new('RGB', (480, 480))

count =0
img2 = tkinter.Label(win)

#实现在本地电脑选择图片
def choose_file():
    select_file = tkinter.filedialog.askopenfilename(title='选择图片')
    e.set(select_file)
    cv_ori = cv2.imread(select_file)
    load = Image.open(select_file)
    load = transforms.Resize((480,480))(load)
    #声明全局变量
    global original
    original = load
    render = ImageTk.PhotoImage(load)
    img  = tkinter.Label(win,image=render)
    img.image = render
    img.place(x=15,y=50)
    
#保存函数
def save():
    global count
    count += 1
    save_img.save(os.path.join(outfile,'test'+str(count)+'.jpg'))

#重置窗口
def refresh():
    temp = original
    new_im = transforms.RandomVerticalFlip(p=1)(temp)
    render = ImageTk.PhotoImage(new_im)
    global img2
    img2.destroy()
    picture1 = tkinter.Label(win,text="原始图片")
    picture1.place(x=15,y=50,width=484,height=484)
    if Checkbutton1_Selected.get()==1 and Checkbutton2_Selected.get()==1:
        Checkbutton1['state'] = 'normal'
        Checkbutton2['state'] = 'normal'
        Checkbutton1.deselect()
        Checkbutton2.deselect()
    
#关闭窗口
def close_window():
    if win:
        try: win.destroy()   
        except (): pass

#随机比例缩放
def lessen():
    temp = original
    new_im = transforms.Resize((100,200))(temp)
    print(f'{temp.size}---->{new_im.size}')
    render = ImageTk.PhotoImage(new_im)
    global img2
    img2.destroy()
    img2  = tkinter.Label(win,image=render)
    img2.image = render
    img2.place(x=560,y=50)
    global save_img
    save_img = new_im

#随机位置裁剪
def random_cut():
    temp = original
    new_im = transforms.RandomCrop(100)(temp)
    render = ImageTk.PhotoImage(new_im)
    global img2
    img2.destroy()
    img2  = tkinter.Label(image=render)
    img2.image = render
    img2.place(x=560,y=50)
    global save_img
    save_img = new_im

#中心剪裁
def center_cut():
    temp = original
    new_im = transforms.CenterCrop(100)(temp)
    render = ImageTk.PhotoImage(new_im)
    global img2
    img2.destroy()
    img2  = tkinter.Label(image=render)
    img2.image = render
    img2.place(x=560,y=50)
    global save_img
    save_img = new_im

#随机水平翻转
def Horizon():
    temp = original
    new_im = transforms.RandomHorizontalFlip(p=1)(temp)
    render = ImageTk.PhotoImage(new_im)
    global img2
    img2.destroy()
    img2  = tkinter.Label(image=render)
    img2.image = render
    img2.place(x=560,y=50)
    global save_img
    save_img = new_im

#随机垂直翻转
def Vertical():
    temp = original
    new_im = transforms.RandomVerticalFlip(p=1)(temp)
    render = ImageTk.PhotoImage(new_im)
    global img2
    img2.destroy()
    img2  = tkinter.Label(image=render)
    img2.image = render
    img2.place(x=560,y=50)
    global save_img
    save_img = new_im

#padding为正方形
def padding():
    temp = original
    new_im = transforms.Pad((0, (temp.size[0]-temp.size[1])//2))(temp)
    render = ImageTk.PhotoImage(new_im)
    global img2
    img2.destroy()
    img2  = tkinter.Label(win,image=render)
    img2.image = render
    img2.place(x=560,y=50)
    global save_img
    save_img = new_im

#随机灰度化
def random_gray():
    temp = original
    new_im = transforms.RandomGrayscale(p=0.5)(temp)
    render = ImageTk.PhotoImage(new_im)
    global img2
    img2.destroy()
    img2  = tkinter.Label(win,image=render)
    img2.image = render
    img2.place(x=560,y=50)
    global save_img
    save_img = new_im
    
#设置对比度
def set_contrast():
    
    def show_contrast(ev=None):
        temp = original
        new_im = transforms.ColorJitter(contrast=scale.get())(temp)
        render = ImageTk.PhotoImage(new_im)
        global img2
        img2.destroy()
        img2  = tkinter.Label(win,image=render)
        img2.image = render
        img2.place(x=560,y=50)
        global save_img
        save_img = new_im

    top = tkinter.Tk()
    top.geometry('250x150')
    top.title('对比度设置')
    scale = tkinter.Scale(top, from_=0, to=100, orient=tkinter.HORIZONTAL,command=show_contrast)
    scale.set(1)
    scale.pack()

#设置色度
def set_hue():
    
    def show_hue(ev=None):
        temp = original
        new_im = transforms.ColorJitter(hue=scale.get())(temp)
        render = ImageTk.PhotoImage(new_im)
        global img2
        img2.destroy()
        img2  = tkinter.Label(win,image=render)
        img2.image = render
        img2.place(x=560,y=50)
        global save_img
        save_img = new_im
        
    top = tkinter.Tk()
    top.geometry('250x150')
    top.title('色度设置')
    scale = tkinter.Scale(top, from_=-0.5, to=0.5, resolution=0.1,orient=tkinter.HORIZONTAL,command=show_hue)
    scale.set(1)
    scale.pack()

#设置饱和度
def set_saturation():
    
    def show_saturation(ev=None):
        temp = original
        new_im = transforms.ColorJitter(saturation=scale.get())(temp)
        render = ImageTk.PhotoImage(new_im)
        global img2
        img2.destroy()
        img2  = tkinter.Label(win,image=render)
        img2.image = render
        img2.place(x=560,y=50)
        global save_img
        save_img = new_im

    top = tkinter.Tk()
    top.geometry('250x150')
    top.title('饱和度设置')
    scale = tkinter.Scale(top, from_=0, to=100, resolution=1,orient=tkinter.HORIZONTAL,command=show_saturation)
    scale.set(1)
    scale.pack()

#设置亮度
def set_bright():

    def show_bright(ev=None):
        temp = original
        new_im = transforms.ColorJitter(brightness=scale1.get(),contrast=scale2.get(), saturation=scale3.get(),hue=scale4.get())(temp)
        render = ImageTk.PhotoImage(new_im)
        global img2
        img2.destroy()
        img2  = tkinter.Label(win,image=render)
        img2.image = render
        img2.place(x=560,y=50)
        global save_img
        save_img = new_im

    top = tkinter.Tk()
    top.geometry('500x300')
    top.title('参数设置')
    scale1 = tkinter.Scale(top, from_=0, to=2,tickinterval = 0.4, resolution =0.05, orient=tkinter.HORIZONTAL,length = 300,command=show_bright)
    scale1.set(0)
    scale1.pack()    
    scale2 = tkinter.Scale(top, from_=0, to=2,tickinterval = 0.4, resolution =0.05, orient=tkinter.HORIZONTAL,length = 300,command=show_bright)
    scale2.set(0)
    scale2.pack() 
    scale3 = tkinter.Scale(top, from_=0, to=2,tickinterval = 0.4, resolution =0.05, orient=tkinter.HORIZONTAL,length = 300,command=show_bright)
    scale3.set(0)
    scale3.pack() 
    scale4 = tkinter.Scale(top, from_=-0.5, to=0.5,tickinterval = 0.2, resolution =0.05, orient=tkinter.HORIZONTAL,length = 300,command=show_bright)
    scale4.set(0)
    scale4.pack() 
    
#中值滤波
def mid_denoise():
    global img2
    global save_img
    if Checkbutton1_Selected.get()==1:
        cvimg = cv2.cvtColor(np.asarray(original),cv2.COLOR_RGB2BGR)
        img_median = cv2.medianBlur(cvimg, 3)
        new_im = Image.fromarray(cv2.cvtColor(img_median,cv2.COLOR_BGR2RGB))
        render = ImageTk.PhotoImage(new_im)
        img2.destroy()
        img2  = tkinter.Label(win,image=render)
        img2.image = render
        img2.place(x=560,y=50)
        save_img = new_im
    else:
        new_im = original
        render = ImageTk.PhotoImage(new_im)
        img2.destroy()
        img2  = tkinter.Label(win,image=render)
        img2.image = render
        img2.place(x=560,y=50)
        save_img = new_im
        
#直方图均衡
def histeq():
    global img2
    global save_img
    if Checkbutton2_Selected.get()==1:
        cvimg = cv2.cvtColor(np.asarray(original),cv2.COLOR_RGB2BGR)
        (b, g, r) = cv2.split(cvimg)
        bH = cv2.equalizeHist(b)
        gH = cv2.equalizeHist(g)
        rH = cv2.equalizeHist(r)
        image = cv2.merge((bH, gH, rH))
        new_im = Image.fromarray(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
        render = ImageTk.PhotoImage(new_im)
        img2.destroy()
        img2  = tkinter.Label(win,image=render)
        img2.image = render
        img2.place(x=560,y=50)
        save_img = new_im
    else:
        new_im = original
        render = ImageTk.PhotoImage(new_im)
        img2.destroy()
        img2  = tkinter.Label(win,image=render)
        img2.image = render
        img2.place(x=560,y=50)
        save_img = new_im
        
#Canny边缘检测
def cannydege():
    if Checkbutton1_Selected.get()==1 and Checkbutton2_Selected.get()==1:
        Checkbutton1['state'] = 'disabled'
        Checkbutton2['state'] = 'disabled'
    cvimg = cv2.cvtColor(np.asarray(original),cv2.COLOR_RGB2GRAY)  # 转成灰度图
    image1 = cv2.GaussianBlur(cvimg,(3,3),0) # 用高斯平滑处理原图像降噪。若效果不好可调节高斯核大小
    image = cv2.Canny(image1, 80, 150)# 指定最大和最小阈值，其中apertureSize默认为3。 (50, 150)
    new_im = Image.fromarray(image)
    render = ImageTk.PhotoImage(new_im)
    global img2
    img2.destroy()
    img2  = tkinter.Label(win,image=render)
    img2.image = render
    img2.place(x=560,y=50)
    global save_img
    save_img = new_im
    
# Hough变换
def houghtrans():
    cvimg = cv2.cvtColor(np.asarray(original),cv2.COLOR_RGB2GRAY) #RGB转灰度
    image1 = cv2.GaussianBlur(cvimg,(3,3),0) #高斯模糊
    edges = cv2.Canny(image1, 50, 90) #边缘检测
    image2 = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR) ##灰度转RGB
    minLineLength = 10 #最小线长阈值
    maxLineGap = 10 #最大线间隙
    lines = cv2.HoughLinesP(edges, 1, np.pi/180.0, 30, minLineLength, maxLineGap)
    # 2,3-直线搜索时的进步尺寸的单位,旋转弧度  4-识别一条直线时它在投票箱中必须达到的值（粗度）
    # 输出为 n*1*4
    if lines is not None:
        lines1 = lines[:,0,:] #提取为二维
        for x1,y1,x2,y2 in lines1[:]:  #画出直线
            cv2.line(image2,(x1, y1), (x2, y2), (0,255,0), 2) #原图，起点，终点，颜色，粗细
    new_im = Image.fromarray(image2)
    render = ImageTk.PhotoImage(new_im)
    global img2
    img2.destroy()
    img2  = tkinter.Label(win,image=render)
    img2.image = render
    img2.place(x=560,y=50)
    global save_img
    save_img = new_im

#随机角度旋转
def Rotation():
    cvimg = cv2.cvtColor(np.asarray(original),cv2.COLOR_RGB2BGR)
    BLACK = [0,0,0]
    image_pad = cv2.copyMakeBorder(cvimg,120,120,120,120,cv2.BORDER_CONSTANT,value=BLACK)
    image = cv2.resize(image_pad,(480,480))#大图为200*200
    pilimg = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    new_im = Image.fromarray(pilimg)
    new_im = transforms.RandomRotation(180)(new_im)
    render = ImageTk.PhotoImage(new_im)
    global img2
    img2.destroy()
    img2  = tkinter.Label(image=render)
    img2.image = render
    img2.place(x=560,y=50)
    global save_img
    save_img = new_im
    
#SIFT
def sift():
    cvoriginal = cv2.cvtColor(np.asarray(original),cv2.COLOR_RGB2BGR)
    
    imgInfo = cvoriginal.shape
    height= imgInfo[0]
    width = imgInfo[1]
    deep = imgInfo[2]
    # 定义一个旋转矩阵(旋转中心，旋转角，缩放比例)
    matRotate = cv2.getRotationMatrix2D((height*0.5, width*0.5), 5, 0.9)
    rot_image = cv2.warpAffine(cvoriginal, matRotate, (height, width))
    rot_image = cv2.GaussianBlur(rot_image,(5,5),0)
    
    cvimg1 = cv2.cvtColor(cvoriginal,cv2.COLOR_RGB2GRAY)
    cvimg2 = cv2.cvtColor(rot_image,cv2.COLOR_RGB2GRAY)
    sift = cv2.xfeatures2d.SIFT_create()

    # FLANN 参数设计
    FLANN_INDEX_KDTREE = 0
    index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
    search_params = dict(checks=50)
    flann = cv2.FlannBasedMatcher(index_params,search_params)
    
    kp1, des1 = sift.detectAndCompute(cvimg1,None) #des是描述子
    kp2, des2 = sift.detectAndCompute(cvimg2,None)

    # FLANN 参数设计2
    matches = flann.knnMatch(des1,des2,k=2)
    matchesMask = [[0,0] for i in range(len(matches))]

    good = []
    for m,n in matches:
        if m.distance < 0.25*n.distance:
            good.append([m])

    image3 = cv2.drawMatchesKnn(cvoriginal,kp1,rot_image,kp2,good,None,flags=2)
    pilimg = cv2.cvtColor(image3,cv2.COLOR_BGR2RGB)
    new_im = Image.fromarray(pilimg)
    render = ImageTk.PhotoImage(new_im)
    global img2
    img2.destroy()
    img2  = tkinter.Label(image=render)
    img2.image = render
    img2.place(x=15,y=50)
    global save_img
    save_img = new_im

#主成分分析
def set_pca():
    
    def Img_PCA(a=80):           #img为输入图像，a为保留主成分比例
        a = float(a)/100
        cvoriginal = cv2.cvtColor(np.asarray(original),cv2.COLOR_RGB2GRAY)
        k = 0
        U,S,V=np.linalg.svd(cvoriginal)  #将img进行奇异值分解，U为左酉矩阵大小M*M，S为奇异值，V为右酉矩阵N*N。
        SS=np.zeros(np.shape(cvoriginal))
        for i in range(len(S)):
            SS[i][i]=S[i]       #将S转化为M*N对角矩阵

        sval=np.sum(S)
        for i in range(len(S)):
            if np.sum(S[:i]) >= a*sval:
                k = i + 1
                break
        Uk=U[:,0:k]
        Sk=SS[0:k,0:k]
        Vk=V[0:k,:]

        pca_img = np.uint8(np.dot(np.dot(Uk,Sk),Vk))  #计算降维处理后的矩阵
        new_im = Image.fromarray(pca_img)
        render = ImageTk.PhotoImage(new_im)
        global img2
        img2.destroy()
        img2  = tkinter.Label(image=render)
        img2.image = render
        img2.place(x=560,y=50)
        global save_img
        save_img = new_im
    top = tkinter.Tk()
    top.geometry('300x150')
    top.title('保留主成分百分比')
    scale = tkinter.Scale(top, from_=0, to=99, resolution=1,tickinterval = 10,length = 260,orient=tkinter.HORIZONTAL,command=Img_PCA)
    scale.set(80)
    scale.pack()

#HOG+SVM目标检测
def ObjDect_HOG():
    
    def is_inside(o, i):
        ox, oy, ow, oh = o
        ix, iy, iw, ih = i
        # 如果符合条件，返回True，否则返回False
        return ox > ix and oy > iy and ox + ow < ix + iw and oy + oh < iy + ih

    # 根据坐标画出人物所在的位置
    def draw_person(image, person):
      x, y, w, h = person
      cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 255), 2)

    # 定义HOG特征+SVM分类器
    cvimg = cv2.cvtColor(np.asarray(original),cv2.COLOR_RGB2BGR)
    hog = cv2.HOGDescriptor()
    hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
    found, w = hog.detectMultiScale(cvimg, winStride=(8, 8), scale=1.05)

    # 判断坐标位置是否有重叠
    found_filtered = []
    for ri, r in enumerate(found):
        for qi, q in enumerate(found):
            a = is_inside(r, q)
            if ri != qi and a:
                break
        else:
            found_filtered.append(r)
    # 勾画筛选后的坐标位置
    for person in found_filtered:
        draw_person(cvimg, person)
    # 显示图像
    pilimg = cv2.cvtColor(cvimg,cv2.COLOR_BGR2RGB)
    new_im = Image.fromarray(pilimg)
    render = ImageTk.PhotoImage(new_im)
    global img2
    img2.destroy()
    img2  = tkinter.Label(image=render)
    img2.image = render
    img2.place(x=560,y=50)
    global save_img
    save_img = new_im
    
# K-means聚类分割
def set_kmeans():
    
    def seg_kmeans_color(k=2):
        k = int(k)
        cvimg = cv2.cvtColor(np.asarray(original),cv2.COLOR_RGB2BGR)
        # 变换一下图像通道bgr->rgb，否则很别扭啊
        b, g, r = cv2.split(cvimg)
        image = cv2.merge([r, g, b])
        # 3个通道展平
        img_flat = image.reshape((image.shape[0] * image.shape[1], 3))
        img_flat = np.float32(img_flat)
        # 迭代参数
        criteria = (cv2.TERM_CRITERIA_EPS + cv2.TermCriteria_MAX_ITER, 20, 0.5)
        flags = cv2.KMEANS_RANDOM_CENTERS
        # 聚类
        compactness, labels, centers = cv2.kmeans(img_flat, k, None, criteria, 6, flags)
        # 显示结果
        img_output = labels.reshape((image.shape[0], image.shape[1]) )
        rate = 256/(k-1)-1  #把类别调成灰度值
        img_output = img_output*rate
        
        new_im = Image.fromarray(img_output)
        render = ImageTk.PhotoImage(new_im)
        global img2
        img2.destroy()
        img2  = tkinter.Label(image=render)
        img2.image = render
        img2.place(x=560,y=50)
        global save_img
        save_img = new_im
    top = tkinter.Tk()
    top.geometry('300x150')
    top.title('设置K的值')
    scale = tkinter.Scale(top, from_=2, to=16, resolution=1,tickinterval = 2,length = 260,orient=tkinter.HORIZONTAL,command=seg_kmeans_color)
    scale.set(2)
    scale.pack()

#CNN图像分割
def xception():
    cvimg = cv2.cvtColor(np.asarray(original),cv2.COLOR_RGB2BGR)
    cv2.imwrite('temp.jpg',cvimg)
    segment_image = semantic_segmentation()
    segment_image.load_pascalvoc_model("deeplabv3_xception_tf_dim_ordering_tf_kernels.h5")
    segment_image.segmentAsPascalvoc("temp.jpg", output_image_name = "out_CNN_seg.jpg", overlay = False)
    new_im = Image.open('out_CNN_seg.jpg')
    render = ImageTk.PhotoImage(new_im)
    global img2
    img2.destroy()
    img2  = tkinter.Label(image=render)
    img2.image = render
    img2.place(x=560,y=50)
    global save_img
    save_img = new_im

#物体识别
def MaskRCNN():
    cvimg = cv2.cvtColor(np.asarray(original),cv2.COLOR_RGB2BGR)
    cv2.imwrite('temp.jpg',cvimg)    
    segment_image = instance_segmentation()
    segment_image.load_model("mask_rcnn_coco.h5")
    segment_image.segmentImage("temp.jpg", output_image_name = "out_MaskRCNN_dect.jpg", show_bboxes = True)
    new_im = Image.open('out_MaskRCNN_dect.jpg')
    render = ImageTk.PhotoImage(new_im)
    global img2
    img2.destroy()
    img2  = tkinter.Label(image=render)
    img2.image = render
    img2.place(x=560,y=50)
    global save_img
    save_img = new_im
    
#显示路径
e = tkinter.StringVar()
e_entry = tkinter.Entry(win, width=80,textvariable=e)
e_entry.place(x=460,y=10)

#设置选择图片的按钮
button1 = tkinter.Button(win, text ="选择图片", command = choose_file)
button1.place(x=0,y=0,width=100,height=40)

#设置退出按钮
button0 = tkinter.Button(win,text="退出系统",command=close_window)
button0.place(x=330,y=0,width=100,height=40)

#设置保存图片的按钮
button2 = tkinter.Button(win,text="保存",command=save)
button2.place(x=110,y=0,width=100,height=40)

#设置标签分别为原图像和修改后的图像
picture1 = tkinter.Label(win,text="原始图片")
picture1.place(x=15,y=50,width=480,height=480)

picture2 = tkinter.Label(win,text="修改后图片")
picture2.place(x=540,y=50,width=480,height=480)

#设置图像去噪
Checkbutton1_Selected = tkinter.IntVar()
Checkbutton1_Selected.set(0)
Checkbutton1 = tkinter.Checkbutton(win,text="图像去噪",variable=Checkbutton1_Selected,onvalue=1,offvalue=0,command=mid_denoise)
Checkbutton1.place(x=47,y=605,width=100,height=30)

#设置直方图均衡化
Checkbutton2_Selected = tkinter.IntVar()
Checkbutton2_Selected.set(0)
Checkbutton2 = tkinter.Checkbutton(win,text="直方图均衡",variable=Checkbutton2_Selected,onvalue=1,offvalue=0,command=histeq)
Checkbutton2.place(x=47,y=655,width=100,height=30)

#设置Hough变换的按钮
button3 = tkinter.Button(win,text="Hough变换",command=houghtrans)
button3.place(x=540,y=600,width=100,height=40)

#设置SIFT的按钮
button4 = tkinter.Button(win,text="SIFT",command=sift)
button4.place(x=860,y=600,width=100,height=40)

#设置PCA降维按钮
button5 = tkinter.Button(win,text="PCA降维",command=set_pca)
button5.place(x=220,y=650,width=100,height=40)

#设置HOG目标检测按钮
button6 = tkinter.Button(win,text="HOG目标检测",command=ObjDect_HOG)
button6.place(x=380,y=650,width=100,height=40)

#设置CNN分割按钮
button7 = tkinter.Button(win,text="CNN分割",command=xception)
button7.place(x=700,y=650,width=100,height=40)

#设置随机角度旋转按钮
button8 = tkinter.Button(win,text="随机旋转",command=Rotation)
button8.place(x=700,y=600,width=100,height=40)

#设置MaskRCNN识别按钮
button9 = tkinter.Button(win,text="MaskRCNN识别",command=MaskRCNN)
button9.place(x=860,y=650,width=100,height=40)

#设置K-means聚类按钮
button9 = tkinter.Button(win,text="K-means聚类",command=set_kmeans)
button9.place(x=540,y=650,width=100,height=40)

#设置参数调节的按钮
button10 = tkinter.Button(win,text="参数调节",command=set_bright)
button10.place(x=220,y=600,width=100,height=40)

#设置Canny边缘检测的按钮
button11 = tkinter.Button(win,text="Canny边缘检测",command=cannydege)
button11.place(x=380,y=600,width=100,height=40)

#设置重置窗口的按钮
button12 = tkinter.Button(win,text="重置窗口",command=refresh)
button12.place(x=220,y=0,width=100,height=40)

win.mainloop()
```

```python
import cv2
import numpy as np
def noise(img,snr):
    h=img.shape[0]
    w=img.shape[1]
    img1=img.copy()
    sp=h*w 
    NP=int(sp*(1-snr))
    for i in range (NP):
        randx=np.random.randint(1,h-1)
        randy=np.random.randint(1,w-1)
        if np.random.random()<=0.5:
            img1[randx,randy]=0
        else:
            img1[randx,randy]=255
    return img1
img = cv2.imread('01.jpg')
img = noise(img,0.98)
cv2.imwrite('00.jpg',img)
```
